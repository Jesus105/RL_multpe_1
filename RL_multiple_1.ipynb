```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#Cargamos el dataset
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

print(boston.DESCR)
```

Formula de minimos cuadrados ordinadios $\alpha = (X^{T}X)^{-1}X^{T}Y$

```python
x1 = np.array(boston.data[:,0])
x2 = np.array(boston.data[:,1])
x3 = np.array(boston.data[:,2])
x4 = np.array(boston.data[:,5])
Y = np.array(boston.target)


X = np.array([np.ones(506), x1, x2, x3, x4]).T


A = np.linalg.inv(X.T @ X) @ X.T @ Y


x_1 = np.array([np.ones(506), x1]).T
x_2 = np.array([np.ones(506), x2]).T
x_3 = np.array([np.ones(506), x3]).T
x_4 = np.array([np.ones(506), x4]).T


A1 = np.linalg.inv(x_1.T @ x_1) @ x_1.T @ Y 
A2 = np.linalg.inv(x_2.T @ x_2) @ x_2.T @ Y 
A3 = np.linalg.inv(x_3.T @ x_3) @ x_3.T @ Y 
A4 = np.linalg.inv(x_4.T @ x_4) @ x_4.T @ Y 
```

```python
print(A)

print(A1)
print(A2)
print(A3)
print(A4)
```

```python
print(A1)
plt.scatter(x1,Y, alpha=0.3)
plt.plot([-1, 80], [A1[0] + A1[1] * -1,A1[0] + A1[1] * 80], c="red")
plt.show()
```

```python
print(A2)
plt.scatter(x2,Y, alpha=0.3)
plt.plot([-1, 100], [A2[0] + A2[1] * -1,A2[0] + A2[1] * 100], c="red")

plt.show()
```

```python
print(A3)
plt.scatter(x3,Y, alpha=0.3)
plt.plot([-1, 30], [A3[0] + A3[1] * -1,A3[0] + A3[1] * 30], c="red")

plt.show()
```

```python
print(A4)
plt.scatter(x4,Y, alpha=0.3)
plt.plot([4, 9], [A4[0] + A4[1] * 4,A4[0] + A4[1] * 9], c="red")

plt.show()
```
